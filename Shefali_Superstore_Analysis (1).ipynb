#Data Cleaning

---



Removing duplicates from the dataset

```python
df.drop_duplicates(inplace=True)
df
```

Dropping unnecessary columns from the dataset

```python
df = df.drop("Country", axis='columns')
df = df.drop("Customer ID", axis='columns')
df = df.drop("Customer Name", axis='columns')
df
```

# Statistical Analysis

---



Total number of items ordered according to each Sub-Category

```python
df.groupby("Sub-Category")["Quantity"].sum()
```

Cost price of each product

```python
df_cost_price = df['Sales'] - df['Profit']
df_cost_price
```

Adding the column of cost price in the dataset

```python
# Adding cost price column in the dataset
df['CostPrice'] = df_cost_price
df
```

Creating a month and year column

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Ship Date"] = pd.to_datetime(df["Ship Date"])

# Creating a month column
df['Month'] = df['Order Date'].dt.strftime('%m')
# Creating a year column
df['Year'] = df['Order Date'].dt.strftime('%Y')
df
```

Number of days each order took to reach the destination

```python
shipping_days = df["Ship Date"] - df["Order Date"]
shipping_days = shipping_days / np.timedelta64(1,'D')
df.insert(loc = 18, column = 'DeliveryDays', value = shipping_days)
df['DeliveryDays'] = df['DeliveryDays'].astype('int')
df
```

## Visualization

---



[ Count of the number of items ordered in each category ]

```python
import matplotlib.pyplot as plt
def category_graph():
  df["Category"].value_counts().plot.bar()
  plt.xlabel("CATEGORY")
  plt.ylabel("COUNT")
  plt.title("Category Count Graph")
category_graph()

#Office supplies are highest in getting ordered compared to other categories.
```

[ Count of the number of items ordered in each sub-category ]

```python
def sub_category_graph():
  df.groupby("Sub-Category")["Quantity"].sum().plot.bar(color="violet")
  plt.xlabel("SUB CATEGORY")
  plt.ylabel("COUNT")
  plt.title("Sub-Category Count Graph")
sub_category_graph()

# Binders are ordered the most in case of each sub-category.
```

[ Count of orders that uses particular ship modes  ]

```python
def ship_mode_graph():
  df["Ship Mode"].value_counts().plot.bar(color = "orange")
  plt.xlabel("SHIP MODE TYPES")
  plt.ylabel("COUNT")
  plt.title("Ship Mode Count Graph")
ship_mode_graph()

# Most of the orders are delivered in standard class ship mode.
```

[ Count of orders collected from the particular states ]

```python
def state_graph():
  df["State"].value_counts().plot.bar(color = "#ff6f68")
  plt.xlabel("STATES")
  plt.ylabel("COUNT")
  plt.title("State Count Graph")
state_graph()

# People from California order the most from this superstore.
```

[ Count of orders collected from different regions ]

```python
def region_graph():
  df["Region"].value_counts().plot.bar(color = "#4dd091")
  plt.xlabel("REGIONS (IN US)")
  plt.ylabel("COUNT")
  plt.title("Region Count Graph")
region_graph()

# Western region of the United States has the highest orders from this store.
```

[ Average Discount added based on each Sub-Category ]

```python
def average_discount_per_subcategory_graph():
  df.groupby("Sub-Category")["Discount"].mean().plot.barh(color="#7db954", label="Average Discount")
  plt.xlabel("AVERAGE DISCOUNT")
  plt.ylabel("SUB-CATEGORY")
  plt.title("Average Discount Per Sub-Category Graph")
  plt.legend()
average_discount_per_subcategory_graph()

# Highest average discount is added to the binders under sub-categories.
```

[ Total Profit/Loss based on each Sub-Category ]

```python
def total_profit_or_loss_per_subcategory_graph():
  df.groupby("Sub-Category")["Profit"].sum().plot.barh(color="#ef70aa", label="Total Profit/Loss")
  plt.xlabel("TOTAL PROFIT/LOSS")
  plt.ylabel("SUB-CATEGORY")
  plt.title("Total Profit/Loss Per Sub-Category Graph")
  plt.legend()
total_profit_or_loss_per_subcategory_graph()

# Copiers acquire higher profit whereas Tables acquire excess loss.
```

[ Total Discount per Segment ]

```python
def total_discount_per_segment_graph():
  df.groupby("Segment")["Discount"].sum().plot.pie(startangle = 120, autopct = "%1.0f%%")
  plt.title("Total Discount per Segment")
total_discount_per_segment_graph()

# In case of total discount, consumers have a greater percentage compared to other segments.
```

[ Total Profit per Region ]

```python
reg = df.groupby("Region")["Profit"].sum().plot.pie(startangle = 90, autopct = "%1.0f%%")
plt.title("Total Profit by Region")

# In case of total profit, larger profit is received from the western region.
```

[ Total Profit by Category ]

```python
def category_profit_pie():
   df.groupby("Category")["Profit"].sum().plot.pie(startangle = 180, autopct = "%1.0f%%")
   plt.title("Total Profit by Category")
category_profit_pie()

# Highest profit is received from Technology as a category.
```

[ Scatter Plot of Product Price and Sales ]

```python
def productprice_by_sales_scatter_plot():
  plt.scatter(df["CostPrice"], df["Sales"], color="#7db954")
  plt.xlabel("PRODUCT PRICE")
  plt.ylabel("SALES")
  plt.title("Scatter Plot of Product Price and Sales")
productprice_by_sales_scatter_plot()

# Most of the pointers lie below 5000 dollars according to the sales and product prices of the products.
```

[ Total Sales per Month ]

```python
def sales_per_month_line_graph():
  df.groupby("Month")["Sales"].sum().plot.line(linestyle="-.", color="#f12761", marker="o", label="Sales")
  plt.xlabel("MONTH")
  plt.ylabel("SALES")
  plt.title("Sales per Month")
  plt.legend()
sales_per_month_line_graph()

# The highest sales were in the month of November(11) and the lowest sales were in the month of February(2).
```

[ Total Profit per Month ]

```python
def profit_per_month_line_graph():
  df.groupby("Month")["Profit"].sum().plot.line(linestyle="-", color="#6883bc", marker="o", label="Profit")
  plt.xlabel("MONTH")
  plt.ylabel("PROFIT")
  plt.title("Profit per Month")
  plt.legend()
profit_per_month_line_graph()

# Highest profit was in the month of December(12) and lowest profit was in the month of January(1).
```

[ Total Sales over the Years ]

```python
def sales_over_the_years_line_graph():
  df.groupby("Year")["Sales"].sum().plot.line(linestyle="-.", color="#d902ee", marker="*", label="Sales")
  plt.xlabel("YEARS")
  plt.ylabel("SALES")
  plt.title("Sales over the Years")
  plt.legend()
sales_over_the_years_line_graph()

# Highest sales were in the year of 2014 and lowest sales were in the year of 2012
```

[ Total Profit over the Years ]

```python
def profit_over_the_years_line_graph():
  df.groupby("Year")["Profit"].sum().plot.line(linestyle="-", color="#097770", marker="*", label="Profit")

  plt.xlabel("YEARS")
  plt.ylabel("PROFIT")
  plt.title("Profit over the Years")
  plt.legend()
profit_over_the_years_line_graph()

# Highest profit was in the year of 2014 and lowest profit was in the year of 2011
```
